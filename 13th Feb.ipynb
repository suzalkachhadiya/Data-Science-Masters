# 2] Write a python program to print Python Exception Hierarchy.

```python
# define a function to print the exception hierarchy
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 2)

# call the function with the base Exception class
print_exception_hierarchy(Exception)
```

### => Using LookupError as a catch-all exception type can be useful when you want to handle any of these types of errors in a single block of code: 
###    1) Indexing out of range
###    2) Key not found
###    3) Name not defined


```python
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']  # Attempt to access a non-existent key
except LookupError:
    print("Key not found!")
```

# 6] List down some best practices for exception handling in python.

## 1) Catch Specific Exceptions: 
### => Catch only the specific exceptions that you can handle, and avoid catching all exceptions. For example, instead of using a generic "except" statement, catch specific exceptions like ValueError, TypeError, etc.


## 2) Use Try-Except Block: 
### => Use the try-except block to handle exceptions. This ensures that your code can handle exceptions gracefully, without crashing.

## 3) 
### => Handle Exceptions as Close to the Source as Possible: Handle exceptions as close to the source as possible, to ensure that you're catching the specific exception that caused the error.

## 4) Provide Detailed Error Messages: 
### => Provide detailed error messages when an exception occurs. This helps users understand what went wrong and how to fix it.

## 5) Use Finally Clause: 
### => Use the "finally" clause to perform cleanup operations, such as closing files or database connections, regardless of whether an exception occurred.

## 6) Avoid Bare Except Clauses: 
### =>Avoid using a bare "except" clause, which catches all exceptions. This can make it difficult to debug your code and can hide potential problems.

## 7) Don't Suppress Exceptions: 
### => Don't suppress exceptions, as this can lead to unexpected behavior or errors. If you can't handle an exception, re-raise it or log it for debugging.

## 8) Use Context Managers:
### => Use context managers, such as "with" statements, to ensure that resources are properly managed and cleaned up when an exception occurs.

## 9) Test Your Exception Handling Code:
### => Test your exception handling code to ensure that it works as expected and that you're handling all possible exceptions.
