# 1] What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

### => Matplotlib is a Python data visualization library that is used to create 2D plots and graphs. It is a powerful tool for generating plots and charts in Python, and it is widely used in the scientific community for data analysis and visualization.
### 
### => It provides a flexible and customizable interface for creating and modifying plots, allowing users to adjust the colors, fonts, and other attributes of their charts to suit their needs.
### 
## 1) Line plot - 
### => a plot that displays data as a series of points connected by lines.

## 2) Scatter plot -
### => a plot that displays data as a collection of points, with each point representing one or more data values.

## 3) Bar plot - 
### => a plot that displays data as a series of bars, with each bar representing a category of data.

## 4) Histogram -
### => a plot that displays the distribution of data values in a dataset.

## 5) Pie chart -
### => a plot that displays data as a set of slices in a circular chart, with each slice representing a portion of the data.

# 2] What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
# Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,c="#FCF105")
plt.title("scatter plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
```

### => The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows users to arrange several subplots in a grid format, with each subplot showing a different aspect of the same data or different data altogether. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
fig,axs=plt.subplots(4)
axs[0].plot(x1,y1,c="r")
axs[1].plot(x2,y2,c="y")
axs[2].plot(x3,y3,c="g")
axs[3].plot(x4,y4,c="b")
plt.show()
```

# 5] What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)

### => A box plot displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median.
### 
### => Box plots are useful for visualizing and comparing the distribution of numerical data between different groups or datasets. They provide information on the range of the data, the median, and the spread of the distribution, as well as any outliers or extreme values. Box plots can also reveal whether a dataset is skewed or symmetric, and whether it contains any significant gaps or clusters.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.show()
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.show()
```
