# 1] What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

## 1) Fixed Acidity - It is the concentration of non-volatile acids present in the wine. The acidity level affects the taste of wine, and it can range from very low to very high.

## 2) Volatile Acidity - This is the concentration of acetic acid in wine, which is produced by the fermentation process. High levels of volatile acidity can cause a vinegary taste, which is undesirable.

## 3) Citric Acid - This is a weak organic acid that is found naturally in many fruits, including grapes. Citric acid is an important indicator of the freshness and quality of wine.

## 4) Residual Sugar - This is the amount of sugar that remains in the wine after fermentation. Wines with higher levels of residual sugar tend to be sweeter.

## 5) Chlorides - These are salts that are present in wine, and their concentration can affect the taste of the wine.

## 6) Free Sulfur Dioxide - This is a preservative that is added to wine to prevent it from spoiling. The concentration of free sulfur dioxide can affect the taste and smell of wine.

## 7) Total Sulfur Dioxide - This is the total amount of sulfur dioxide in wine, including both free and bound forms.

## 8) Density - The density of wine can provide information about its alcohol content, sweetness, and overall quality.

## 9) pH - The pH of wine is a measure of its acidity, and it can range from 2.8 to 4.0.

## 10) Sulphates - These are compounds that are added to wine to prevent oxidation and spoilage. The concentration of sulphates can affect the taste and smell of wine.

## 11) Alcohol - The alcohol content of wine can range from 8% to 15%. It affects the overall taste and quality of wine.

## 12) Quality - This is the target variable, which ranges from 0 to 10 and is based on sensory data that rates the quality of the wine.

# 6] Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

wine_data = pd.read_csv('winequality-red (1).csv')

X = wine_data.drop('quality', axis=1)
y = wine_data['quality']

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# perform PCA
pca = PCA()
pca.fit(X_std)

explained_var_ratio = pca.explained_variance_ratio_

cumulative_expl_var_ratio = np.cumsum(explained_var_ratio)

n_components = np.argmax(cumulative_expl_var_ratio >= 0.9) + 1

print("Minimum number of principal components required to explain 90% of variance:", n_components)
```
