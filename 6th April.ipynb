### f(x) => represents the predicted class for a given data point x.

### w => represents the weight vector perpendicular to the hyperplane.
### · => denotes the dot product between two vectors.
### x => represents the feature vector of a data point.
### b => is the bias term or the intercept of the hyperplane.

# 2] What is the objective function of a linear SVM?


### => The objective function of a linear Support Vector Machine (SVM) is formulated to maximize the margin between the classes while minimizing the classification error. The most commonly used formulation for linear SVM is the hinge loss function.

### => The objective function of a linear SVM can be defined as:

### minimize 1/2 ||w||² + C Σ[max(0, 1 - yᵢ(w • xᵢ + b))]

### subject to yᵢ(w • xᵢ + b) ≥ 1 for all training samples (xᵢ, yᵢ)

# 3] What is the kernel trick in SVM?


### => The kernel trick is a technique used in Support Vector Machines (SVMs) to handle nonlinearly separable data by implicitly mapping the input features into a higher-dimensional feature space. It allows SVMs to operate in a high-dimensional space without explicitly computing the transformed features.

### => The idea behind the kernel trick is to define a kernel function that computes the dot product between two vectors in the high-dimensional feature space, without explicitly transforming the original input vectors. This avoids the computational cost and potential memory constraints of explicitly working in high-dimensional spaces.
### Some commonly used kernel functions include:

## 1) Linear kernel:
### K(x, x') = x • x'
### => This is equivalent to the original linear SVM formulation, where the dot product is computed in the original feature space.

## 2) Polynomial kernel: 
### K(x, x') = (γ(x • x') + r)^d
### => This kernel applies a polynomial transformation to the original features and computes the dot product in the transformed space. γ, r, and d are kernel parameters.

### 3) Radial Basis Function (RBF) kernel:
### K(x, x') = exp(-γ||x - x'||²)
### => The RBF kernel measures the similarity between two samples based on their Euclidean distance in the original feature space. γ is a kernel parameter controlling the width of the Gaussian function.

# 4] What is the role of support vectors in SVM Explain with example


### => In a Support Vector Machine (SVM), support vectors play a crucial role in defining the decision boundary and determining the classification outcome. They are the subset of training samples that lie closest to the decision boundary or are misclassified. The support vectors provide the necessary information for constructing the decision boundary and optimizing the SVM model.

### => To explain the role of support vectors, let's consider a binary classification example with two classes: positive examples represented by blue dots and negative examples represented by red dots. The goal is to find a linear decision boundary that separates the two classes.

### => In the case of a linear SVM, the decision boundary is a hyperplane defined by the equation w • x + b = 0, where w is the weight vector perpendicular to the hyperplane, x represents the input vector, and b is the bias term.

### => During the training process, the SVM algorithm identifies the support vectors, which are the samples lying closest to the decision boundary or within the margin region. These support vectors have a significant influence on determining the position and orientation of the decision boundary.

### => In the example, let's say the three circled samples (two blue and one red) are identified as the support vectors. These support vectors lie either on the margin or closer to the decision boundary.

### => By using the support vectors, the SVM algorithm finds the optimal weight vector w and bias term b that minimize the classification error and maximize the margin. The support vectors contribute to the objective function of the SVM and affect the final decision boundary.

### => The support vectors define the margin of the SVM, which is the region separating the two classes. The margin is determined by the distance between the decision boundary and the support vectors. Maximizing the margin helps achieve better generalization and robustness of the SVM model.

### => During the prediction phase, the support vectors are essential for determining the class label of new, unseen samples. The decision function of the SVM relies on the support vectors' information and assigns the class label based on their influence.

# 5] Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?


## 1) Hyperplane:
### => In SVM, a hyperplane is a decision boundary that separates data points of different classes in a classification problem. In a binary classification scenario, the hyperplane is a (d-1)-dimensional subspace in a d-dimensional feature space. It can be represented by the equation:
### w⋅x + b = 0

### => where w represents the normal vector to the hyperplane, x is the input feature vector, and b is the bias term.

## 2) Marginal Plane:
### => The marginal plane refers to the hyperplanes parallel to the decision boundary (hyperplane) that are closest to the support vectors. These planes play a crucial role in determining the margin of the SVM classifier.

## 3) Soft Margin:
### => Soft margin SVM allows for misclassified points and slight violations of the margin. It introduces a slack variable (ξ) for each training example, which represents the degree of misclassification or margin violation. The objective is to minimize both the margin and the sum of slack variables.

## 4) Hard Margin:
### => Hard margin SVM enforces strict separation of the classes without allowing any misclassification or margin violation. It assumes that the data points are linearly separable. In this case, the margin is maximized, and no slack variables are allowed.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Load the Iris dataset
iris = load_iris()

# Split the dataset into features (X) and target labels (y)
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=2)

# Create an SVM classifier
svm = SVC()

# Train the classifier
svm.fit(X_train, y_train)

# Make predictions on the test set
predictions = svm.predict(X_test)
```

```python
# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
accuracy2 = accuracy_score(y_test, predict2)
print("Accuracy:", accuracy2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC

# Load the Iris dataset
iris = load_iris()

# Consider two features for visualization
X = iris.data[:, :2]  # Selecting the first two features
y = iris.target

# Train an SVM classifier
svm = SVC()
svm.fit(X, y)

# Create a meshgrid to plot the decision regions
h = 0.02  # Step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries and data points
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundaries of SVM')
plt.colorbar()

plt.show()
```

## Implement a linear SVM classifier from scratch using Python and compare its performance with the scikit-learn implementation.

```python
class LinearSVM:
    def __init__(self, learning_rate=0.01, lambda_param=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Gradient descent optimization
        for _ in range(self.num_iterations):
            scores = np.dot(X, self.weights) + self.bias
            margins = y * scores

            # Misclassified samples
            misclassified = np.where(margins < 1)[0]

            # Update weights and bias
            self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights - np.dot(X.T, y))
            self.bias -= self.learning_rate * np.sum(-2 * self.lambda_param * self.weights)

    def predict(self, X):
        scores = np.dot(X, self.weights) + self.bias
        return np.sign(scores)
```

```python
# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.19, random_state=2)

# Create and train the linear SVM classifier from scratch
svm_scratch = LinearSVM()
svm_scratch.fit(X_train, y_train)

# Make predictions using the linear SVM classifier from scratch
predictions_scratch = svm_scratch.predict(X_test)

# Calculate the accuracy of the linear SVM classifier from scratch
accuracy_scratch = accuracy_score(y_test, predictions_scratch)
print("Accuracy of SVM classifier from scratch:", accuracy_scratch)
```
